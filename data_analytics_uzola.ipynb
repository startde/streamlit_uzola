# Тестирование четырех входо модуле1 1,3,4

```python
df = pd.read_excel("/content/signals1-4.xlsx")
#df = df.rename(columns={"                     N-входов       \n\nСкорость (кГц)\n                                        ": "N-входов"})
```

```python
df['ID модуля.номер входа'] = df['ID модуля.номер входа'].apply(lambda x: f"{x:.1f}")
```

```python
#df['Отклонение от заданного числа импульсов в %'] = df['Отклонение от заданного числа импульсов в %'].apply(lambda x: f"{x:.5f}")
```

```python
df
```

```python
table_1 = df['Отклонение от заданного числа импульсов в %']
type(table_1)
```

```python
for i in table_1:
  print (type(i))
```

```python
arr = np.array_split(table_1, 3)
arr
```

```python
plt.hist(arr[2])
```

```python
di_1,di_3,di_4 = arr[0], arr[1], arr[2]
di_1.tolist()
di_3.tolist()
di_4.tolist()
di_1 = [float(i) for i in di_1]
di_3 = [float(i) for i in di_3]
di_4 = [float(i) for i in di_4]
```

```python
print('{:0.9f}'.format(statistics.mean(di_4)))
```

```python
modules_mean = {"di_1":statistics.mean(di_1) , "di_3":statistics.mean(di_3), "di_4": statistics.mean(di_4) }
modules_mean
```

```python
plt.bar(list(modules_mean.keys()),
        list(modules_mean.values()),
        color ='green',
        width = 0.4)
plt.title ("Среднее отклонение по модулям")
```
